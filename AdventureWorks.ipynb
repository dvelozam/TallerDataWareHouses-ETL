```python
###Carga de datos a BigQuery
```

```python
credentials = service_account.Credentials.from_service_account_file( 
    '/Users/diego/Downloads/adventureworks-315200-c8d9cedc4faa.json', 
    scopes = [ 'https://www.googleapis.com/auth/cloud-platform' ]
)
```

```python
client = bigquery.Client( credentials = credentials, project = credentials.project_id )
```

```python
job_config = bigquery.LoadJobConfig(
    schema = [
        # Supported datatypes: https://cloud.google.com/bigquery/docs/reference/standard-sql/data-types
        bigquery.SchemaField( 'firstname', bigquery.enums.SqlTypeNames.STRING),
        bigquery.SchemaField( 'jobtitle', bigquery.enums.SqlTypeNames.STRING),
        bigquery.SchemaField( 'startdate', bigquery.enums.SqlTypeNames.DATE),
        bigquery.SchemaField( 'name', bigquery.enums.SqlTypeNames.STRING),
    ],
    # Drod and re-create table, if exist
    write_disposition = 'WRITE_TRUNCATE',
)
```

```python
job = client.load_table_from_dataframe(
    table4, 'DB2AdventureWorks.table4', job_config = job_config
)
job.result()
```

```python
table = client.get_table( 'DB2AdventureWorks.table4' )
print(
    'Loaded {} rows and {} columns to {}'.format(
        table.num_rows, len( table.schema ), 'DB2AdventureWorks.table4'
    )
)
```
