```python
import psycopg2
import pandas as pd

from google.cloud import bigquery
from google.oauth2 import service_account
```

# Segundo Punto

```python
conn = psycopg2.connect( host = 'localhost', port = 5432, database = 'AdventureWorks', user = 'postgres',  password="1234")
```

<span style="color:red">¿Cuántas bicicletas hay por subcategoría?</span>

```python
cur = conn.cursor()
consul1 = pd.read_sql_query( 'SELECT production.productcategory.Name, production.product.Name, count(production.productcategory.productcategoryid) AS ID_CATEGORIAS FROM production.product, production.productcategory, production.productsubcategory WHERE production.product.ProductSubcategoryID = production.productsubcategory.ProductSubcategoryID AND production.productsubcategory.ProductCategoryID = production.productcategory.ProductCategoryID GROUP BY ROLLUP(production.productcategory.productcategoryid, production.productcategory.Name, production.product.Name)', conn )
cur.close()
```

```python
consul1.head(20)
```

 <span style="color:green">Respuesta: hay 97 bicicletas por cada subcategoría.</span>

 <span style="color:red">¿Cuántos productos se ofrecen en la categoría Accessories?</span>

```python
cur = conn.cursor()
consul2 = pd.read_sql_query( 'SELECT  production.productcategory.Name, count(production.PRODUCT.productID) AS Productos FROM production.product, production.productcategory, production.productsubcategory WHERE production.product.ProductSubCategoryID = production.productsubcategory.ProductSubCategoryID AND production.productsubcategory.ProductCategoryID = production.productcategory.ProductCategoryID GROUP BY ROLLUP(production.productcategory.Name, production.PRODUCT.productID )', conn )
cur.close()
```

```python
consul2.head(20)
```

<span style="color:green">Respuesta: se ofrecen 29 productos en la categoría de Accesorios ‘Accessories’. </span>

<span style="color:red">¿Cuál es el rango de fecha de las órdenes programadas por la organización empresarial para su área de producción? </span>


```python
cur = conn.cursor()
consul3 = pd.read_sql_query( 'SELECT production.workorderrouting.ScheduledStartDate, production.workorderrouting.ScheduledEndDate FROM  production.workorderrouting GROUP BY (production.workorderrouting.ScheduledStartDate, production.workorderrouting.ScheduledEndDate) ORDER BY (production.workorderrouting.ScheduledStartDate)', conn )
cur.close()
```

```python
consul3.head(20)
```

<span style="color:green">Respuetas: los rangos de fechas de las órdenes programadas por el área de producción van desde el día tres del mes de junio del año 2011 hasta el día dos del mes de junio del año 2014.</span>

<span style="color:red">¿Con cuántos vendedores se mantiene el área de adquisiciones de la empresa?</span>

```python
cur = conn.cursor()
consul4 = pd.read_sql_query( 'SELECT purchasing.vendor.Name, purchasing.vendor.businessentityid, Count(purchasing.vendor.Name) as count_vendedor FROM purchasing.vendor GROUP BY ROLLUP(purchasing.vendor.Name, purchasing.vendor.businessentityid)', conn )
cur.close()
```

```python
consul4.head(20)
```

<span style="color:green"> Respuesta: existen 104 vendedores con los cuales, la empresa mantiene contacto en el área de adquisición. </span>

<span style="color:red"> ¿Cuáles y cuántos son los territorios en los que la empresa tiene sus tiendas? </span>

```python
cur = conn.cursor()
consul5 = pd.read_sql_query( 'SELECT sales.salesterritory.Name, count(sales.salesterritory.Name ) As CANTIDAD FROM sales.salesterritory GROUP BY ROLLUP(sales.salesterritory.Name )', conn )
cur.close()
```

```python
consul5.head(20)
```

<span style="color:green"> Respuesta: la empresa tiene sus tiendas en 10 territorios, los cuales son: Francia, el sur oeste, Reino Unido, noroeste, central, noreste, Australia, Alemania, sureste, Canadá. </span>

# Cuarto Punto

<span style="color:blue"> Carga de datos a BigQuery </span>

```python
credentials = service_account.Credentials.from_service_account_file( 
    '/Users/diego/OneDrive/Escritorio/Bases de datos 2/AdventureWorks/adventureworks-315200-c8d9cedc4faa.json', 
    scopes = [ 'https://www.googleapis.com/auth/cloud-platform' ]
)
```

```python
client = bigquery.Client( credentials = credentials, project = credentials.project_id )
```

```python
job_config = bigquery.LoadJobConfig(
    schema = [
        # Supported datatypes: https://cloud.google.com/bigquery/docs/reference/standard-sql/data-types
        bigquery.SchemaField( 'firstname', bigquery.enums.SqlTypeNames.STRING),
        bigquery.SchemaField( 'jobtitle', bigquery.enums.SqlTypeNames.STRING),
        bigquery.SchemaField( 'startdate', bigquery.enums.SqlTypeNames.DATE),
        bigquery.SchemaField( 'name', bigquery.enums.SqlTypeNames.STRING),
    ],
    # Drod and re-create table, if exist
    write_disposition = 'WRITE_TRUNCATE',
)
```

```python
job = client.load_table_from_dataframe(
    table4, 'DB2AdventureWorks.table4', job_config = job_config
)
job.result()
```

```python
table = client.get_table( 'DB2AdventureWorks.table4' )
print(
    'Loaded {} rows and {} columns to {}'.format(
        table.num_rows, len( table.schema ), 'DB2AdventureWorks.table4'
    )
)
```
